```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
loki=cv2.imread('loki.png')
loki=cv2.cvtColor(loki,cv2.COLOR_BGR2RGB)
```

```python
lokig=cv2.imread('lokig.png')
lokig=cv2.cvtColor(lokig,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(loki)
```

```python
plt.imshow(lokig)
```

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
  face_img=img.copy()
  face_rects=face_cascade.detectMultiScale(face_img)
  for (x,y,w,h) in face_rects:
    cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
  return face_img
```

```python
result=detect_face(loki)
```

```python
plt.imshow(result)
```

```python
def adj_detect_face(img):

    face_img = img.copy()

    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2, minNeighbors=3)

    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10)

    return face_img
```

```python
adj_detect_face(loki)
```

```python
adj_detect_face(lokig)
```

```python
detect_face(lokig)
```
