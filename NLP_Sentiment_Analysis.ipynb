```python
import numpy as np
import pandas as pd
import nltk
```

```python
df=pd.read_csv("harry_potter_reviews.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
new_data=df[['comment']]
```

```python
len(new_data)
```

```python
df.dropna(inplace=True)
blanks=[]
for i,cm in new_data.itertuples():
  if type(cm)==str:
    if cm.isspace():
      blanks.append(i)
df.drop(blanks,inplace=True)
```

```python
len(new_data)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
sid.polarity_scores(new_data.loc[0]['comment'])
```

```python
new_data.loc[0]['comment']
```

```python
new_data['scores']=df['comment'].apply(lambda comment:sid.polarity_scores(comment))
```

```python
new_data.head(10)
```

```python
new_data['compound']  = new_data['scores'].apply(lambda score_dict: score_dict['compound'])

new_data.head()
```

```python
new_data['comp_score'] = new_data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')

new_data.head()
```

```python
new_data[['comment','comp_score']]
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = new_data['comment']
y = new_data['comp_score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC

text_clf = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', LinearSVC()),
])

# Feed the training data through the pipeline
text_clf.fit(X_train, y_train)
```

```python
predictions = text_clf.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,predictions))
```

```python
print(metrics.classification_report(y_test,predictions))
```

```python
print(metrics.accuracy_score(y_test,predictions))
```
